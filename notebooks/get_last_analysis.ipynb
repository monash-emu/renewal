```python
import os
from datetime import datetime

from emu_renewal.inputs import OUTPUTS_PATH
```

```python
def get_latest_analyses(country, analyses, date_format="%Y%m%d_%H%M"):
    last_analyses = {}
    for analysis in analyses:
        path = OUTPUTS_PATH / country / analysis
        dates = [datetime.strptime(d, date_format) for d in os.listdir(path)]
        last_analyses[analysis] = datetime.strftime(max(dates), date_format)
    return last_analyses
```

```python
get_latest_analyses("France", ["fb_linear", "fb_square", "no_mob"])
```
