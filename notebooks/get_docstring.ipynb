```python
import re
from emu_renewal.inputs import get_worldbank_national_pop
```

```python
def get_func_blurb(function):
    docstring = function.__doc__
    blurb = re.split("(Args|Returns):", docstring)[0]
    return re.sub(r"\s+", " ", blurb)
```
