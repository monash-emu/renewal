```python
from os import listdir as ls
from matplotlib import pyplot as plt
from emu_renewal.constants import OUTPUTS_PATH
from emu_renewal.inputs import get_world_shp
from emu_renewal.outputs import get_median_ratios, get_param_vals_by_analysis, get_ratios_from_disps
plt.style.use("default")
```

```python
job_path = OUTPUTS_PATH / "46693102"
all_countries = ls(job_path)
disp_posts = {c: get_param_vals_by_analysis("dispersion_proc", job_path / c) for c in all_countries}
ratio_dists = get_ratios_from_disps(disp_posts)
```

```python
world = get_world_shp()
fig, ax = plt.subplots(1, 1, figsize=[15, 12], constrained_layout=True)
median_ratios = get_median_ratios(ratio_dists, "g_mob")

world["disp_ratio"] = world["ISO_A3"].map(median_ratios)
mob_avail = world[world["disp_ratio"].notna()]
mob_unavail = world[world["disp_ratio"].isna()]

mob_avail.plot(column="disp_ratio", ax=ax, cmap="RdGy_r", legend=True, vmin=0.4, vmax=1.6)
mob_unavail.plot(ax=ax, color="w", hatch="///", edgecolor="whitesmoke")
world.boundary.plot(ax=ax, color="black", linewidth=0.2)

world["centroid"] = world.geometry.centroid
centroids = world.set_geometry("centroid")
centroids.plot(ax=ax, color="black", markersize=30)

ax.set_xticks([])
ax.set_yticks([]);
```
