```python
import json

from emu_renewal.constants import DATA_PATH, RERUNS, FULL_RUN, TIMEOUTS, ANALYSIS_NAMES
from emu_renewal.utils import get_analysis_paths
from emu_renewal.outputs import get_param_vals_by_analysis, get_ratios_from_disps
from emu_renewal.plotting import plot_exponent_dispersion_comparison_interactive
```

```python
all_countries = json.load(open(DATA_PATH / "config/included.json", "r"))
analysis_paths = get_analysis_paths(RERUNS + FULL_RUN + TIMEOUTS, all_countries)
```

```python
disp_posts = {c: get_param_vals_by_analysis("dispersion_proc", path) for c, path in analysis_paths.items()}
ratio_dists = get_ratios_from_disps(disp_posts)
```

```python
for mob_source in list(ANALYSIS_NAMES.keys())[1: -1]:
    fig = plot_exponent_dispersion_comparison_interactive(analysis_paths, mob_source, ratio_dists)
    fig.write_html(f"exp_disp_comparison_{mob_source}.html")
```
