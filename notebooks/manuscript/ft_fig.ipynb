```python
from pathlib import Path
import os
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import subplots
from matplotlib_inline.backend_inline import set_matplotlib_formats
import warnings

from emu_renewal.constants import OUTPUTS_PATH, MOB_SOURCE_COLOURS, MOB_SOURCE_ABBREVS
from emu_renewal.outputs import load_targets
from emu_renewal.plotting import plot_composite_calibrations

set_matplotlib_formats("svg")
```

```python
job_path = OUTPUTS_PATH / "47213985"
iso3 = "FRA"
c_path = job_path / iso3
analyses = {d.name: Path(d.path) for d in os.scandir(job_path / iso3) if d.is_dir()}
c_procs = [pd.read_hdf(c_path / a / "spaghetti.h5")["process"] for a in analyses]
procs = pd.concat(c_procs, keys=analyses, axis=1)
```

```python
fig, ax = subplots(1, 1, figsize=[4, 3], sharey=True)
ax.set_yticks([])
ax.tick_params(axis="x", labelrotation=45)
ans = [a for a in MOB_SOURCE_ABBREVS if a in analyses]
for a in ans:
    colour = MOB_SOURCE_COLOURS[a]
    label = MOB_SOURCE_ABBREVS[a].replace("FB", "Facebook").replace("none", "no scaling for mobility")
    quants = procs[a].quantile([0.05, 0.5, 0.95], axis=1).T
    ax.plot(quants.index, quants[0.5], color=colour, label=label, linewidth=2.0)
    ax.fill_between(quants.index, quants[0.05], quants[0.95], alpha=0.1, color=colour)
ax.legend();
```
