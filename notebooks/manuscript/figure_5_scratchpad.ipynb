```python
from matplotlib import pyplot as plt
import arviz as az
import seaborn as sns

from emu_renewal.constants import OUTPUTS_PATH
from emu_renewal.inputs import get_google_mobility
from emu_renewal.plotting import G_MOB_DOMAIN_CMAP
```

```python
iso3 = "AUT"
job_path = OUTPUTS_PATH / "46090067"
idata = az.from_netcdf(job_path / iso3 / "g_mob/idata_filtered.nc")
weights = idata.posterior["mob_weights"].to_dataframe().unstack("mob_weights_dim_0")
mob = get_google_mobility(iso3)
weights.columns = mob.columns
fig, ax = plt.subplots(1, 2, figsize=[10, 5])
sns.kdeplot(weights, fill=True, alpha=0.05, linewidth=1, ax=ax[0], palette=G_MOB_DOMAIN_CMAP)
```
