```python
from os import listdir as ls
from IPython.display import display, Markdown
import pandas as pd
import pycountry
import seaborn as sns

from emu_renewal.inputs import OUTPUTS_PATH, DATA_PATH, get_gdps, get_worldbank_national_pop
from emu_renewal.plotting import ANALYSIS_NAMES, get_idatas_for_mob_type, get_param_quant_descript
```

```python
job_path = OUTPUTS_PATH / "44255911"
countries = ls(job_path)
all_countries = ls(job_path)
mob_types = ["g_mob", "fb_mob", "a_mob"]
```

```python
quants = {}
for mob_type in mob_types:
    idatas, unavailable = get_idatas_for_mob_type(job_path, all_countries, mob_type)
    quants[mob_type] = get_param_quant_descript("mob_exp", idatas, 3)
results = pd.DataFrame({ANALYSIS_NAMES[mob]: quants[mob] for mob in mob_types})
quants["gdp"] = get_gdps(2020)
quants_df = pd.DataFrame(quants)
```

```python
quants_df["pop_size"] = quants_df.index.map(lambda x: get_worldbank_national_pop(x, 2020))
```

```python
sns.scatterplot(x="gdp", y="a_mob", size="pop_size", data=quants_df, sizes=(40, 2000))
```
