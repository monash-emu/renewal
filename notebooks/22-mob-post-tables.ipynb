```python
from os import listdir as ls
from IPython.display import display, Markdown
import pandas as pd
import pycountry

from emu_renewal.inputs import OUTPUTS_PATH
from emu_renewal.plotting import ANALYSIS_NAMES, get_idatas_for_mob_type
```

```python
job_path = OUTPUTS_PATH / "44255911"
countries = ls(job_path)
all_countries = ls(job_path)
mob_types = ["g_mob", "a_mob"]
```

```python
def get_param_quant_descript(idatas, places):
    out = {}
    for c in idatas:
        country = pycountry.countries.lookup(c).name
        quants = idatas[c].posterior["mob_exp"].quantile([0.5, 0.05, 0.95]).data
        text = f"{round(quants[0], places)} ({round(quants[1], places)} to {round(quants[2], places)})"
        out[country] = text
    return out
```

```python
quants = {}
for mob_type in mob_types:
    idatas, unavailable = get_idatas_for_mob_type(job_path, all_countries, mob_type)
    quants[mob_type] = get_param_quant_descript(idatas, 3)
pd.DataFrame({ANALYSIS_NAMES[mob]: quants[mob] for mob in mob_types}).fillna("Mobility unavailable")
```
