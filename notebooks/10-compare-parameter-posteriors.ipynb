```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
from matplotlib.pyplot import cm
from os import listdir as ls

from emu_renewal.inputs import OUTPUTS_PATH, get_standard_priors
from emu_renewal.outputs import get_param_vals_by_analysis
from emu_renewal.plotting import plot_kde_comparison, get_param_medians, AN_ABBREVS
from emu_renewal.utils import group_countries_by_continent, get_countries_by_continent
```

```python
colours = cm.Dark2.colors
```

```python
job_path = OUTPUTS_PATH / "44255911"
countries = ls(job_path)
all_countries = ls(job_path)
countries_by_cont = get_countries_by_continent(all_countries)
```

```python
temp_countries = {"NA": countries_by_cont["NA"]}
```

```python
for cont, countries in temp_countries.items():
    param_posts = {}
    for country in countries:
        vals = get_param_vals_by_analysis("dispersion_proc", job_path / country)
        param_posts[country] = vals#.rename(columns=AN_ABBREVS)
    plot_kde_comparison(param_posts, colours, cont)
```
