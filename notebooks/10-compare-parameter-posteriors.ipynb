```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
from matplotlib.pyplot import cm
from os import listdir as ls

from emu_renewal.inputs import OUTPUTS_PATH, get_standard_priors
from emu_renewal.outputs import get_param_vals_by_analysis
from emu_renewal.plotting import plot_kde_comparison, get_param_medians
from emu_renewal.utils import group_countries_by_continent
```

```python
set_matplotlib_formats("svg")
colours = cm.Dark2.colors
```

```python
job_path = OUTPUTS_PATH / "43514097"
countries = ls(job_path)
```

```python
continents_map = group_countries_by_continent(countries)
continents_map["EU1"] = continents_map["EU"][:16]
continents_map["EU2"] = continents_map["EU"][16:]
del continents_map["EU"]
```

```python
label_map = {
    "no_mob": "none",
    "g_mob": "g",
    "fb_mob": "fb",
    "a_mob": "a",
}
```

```python
for cont, countries in continents_map.items():
    param_posts = {}
    for param in ["dispersion_proc"]:
        param_posts[param] = {}
        for country in countries:
            country_path = job_path / country
            vals = get_param_vals_by_analysis(param, country_path)
            param_posts[param][country] = vals.rename(columns=label_map)
    plot_kde_comparison(param_posts["dispersion_proc"], colours, cont, "disp")
```
