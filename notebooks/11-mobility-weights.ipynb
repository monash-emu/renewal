```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
from matplotlib.pyplot import cm
from os import listdir as ls
from IPython.display import display, Markdown

from emu_renewal.inputs import OUTPUTS_PATH, get_google_mobility, get_apple_mobility
from emu_renewal.plotting import plot_mob_weights_by_country
from emu_renewal.utils import get_countries_by_continent
```

```python
set_matplotlib_formats("svg")
```

```python
job_path = OUTPUTS_PATH / "44255911"
countries = ls(job_path)
all_countries = ls(job_path)
countries_by_cont = get_countries_by_continent(all_countries)
```

```python
def get_cont_mobility(cont, mob_type):
    mob = {}
    if mob_type == "g_mob":
        get_mob = get_google_mobility
    elif mob_type == "a_mob":
        get_mob = get_apple_mobility
    for c in countries_by_cont[cont]:
        try:
            mob[c] = get_mob(c)
        except:
            print(f"no {mob_type} data available for {c}")
    return mob
```

```python
for cont in countries_by_cont:
    mob = get_cont_mobility(cont, "g_mob")
    if cont == "EU":
        eur_countries = list(mob.keys())
        display(plot_mob_weights_by_country(job_path, "g_mob", {k: mob[k] for k in eur_countries[:16]}))
        display(plot_mob_weights_by_country(job_path, "g_mob", {k: mob[k] for k in eur_countries[16:]}))
    else:
        display(plot_mob_weights_by_country(job_path, "g_mob", mob))
```
