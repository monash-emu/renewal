```python
import pandas as pd
pd.options.plotting.backend = "plotly"
import pycountry
import pycountry_convert as pc
from datetime import datetime
from emu_renewal.inputs import get_var_target, get_prealpha_vars, \
    get_country_vars, get_delta_vars, get_continent_prealpha_vars, get_continent_data, \
    get_specific_var_props, get_aust_ba2_prop, get_pooled_totals
```

```python
country = "India"
pycountry.countries.lookup(country)
iso3 = pycountry.countries.lookup(country).alpha_3
```

```python
# Prealpha profile
iso2 = pycountry.countries.lookup(iso3).alpha_2
continent = pc.country_alpha2_to_continent_code(iso2)
prealpha_vars = get_prealpha_vars(iso3)
if prealpha_vars is None:
    cont_data = get_continent_data(continent, "prealpha")
    prealpha_vars = get_continent_prealpha_vars(cont_data)
    prealpha_prop = get_pooled_totals(prealpha_vars)["prealpha_prop"]
```

```python
prealpha_prop.plot()
```

```python
delta_data = get_delta_vars(iso3)
date_mask = (datetime(2021, 2, 1) < delta_data.index) & (delta_data.index < datetime(2021, 9, 1))
delta_data.loc[date_mask, "delta_prop"].plot()
```
