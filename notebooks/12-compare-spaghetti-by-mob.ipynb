```python
from emu_renewal.outputs import get_output_dir
import pandas as pd
import plotly.graph_objects as go
```

```python
country = "France"
no_mob_time = "20250114_1642"
mob_time = "20250114_1638"
```

```python
no_mob_spaghetti = pd.read_hdf(get_output_dir(country, "no_mob", no_mob_time) / "spaghetti.h5")
mob_spaghetti = pd.read_hdf(get_output_dir(country, "mob", mob_time) / "spaghetti.h5")
proc_comp_df = pd.concat([no_mob_spaghetti["process"], mob_spaghetti["process"]], keys=["no_mob", "mob"], axis=1)
```

```python
proc_comparison_fig = go.Figure()
line_style = {"color": "orange", "width": 0.2}
spagh = no_mob_spaghetti["process"]
for run in spagh:
    proc_comparison_fig.add_trace(go.Scatter(x=spagh.index, y=spagh[run], line=line_style))
line_style = line_style | {"color": "blue"}
spagh = mob_spaghetti["process"]
for run in spagh:
    proc_comparison_fig.add_trace(go.Scatter(x=spagh.index, y=spagh[run], line=line_style))
proc_comparison_fig.update_layout(showlegend=False, height=500, width=800)
```
