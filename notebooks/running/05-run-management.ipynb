```python
import json
import pandas as pd
from emu_renewal.constants import DATA_PATH, OUTPUTS_PATH, ANALYSIS_TYPES
```

```python
def get_analysis_status(run_id):
    countries = json.load(open(DATA_PATH / "config/included.json", "r"))
    analysis_status = pd.DataFrame(index=countries, columns=ANALYSIS_TYPES)
    run_path = OUTPUTS_PATH / run_id
    for iso3 in countries:
        c_path = run_path / iso3
        for a in ANALYSIS_TYPES:
            analysis_status.loc[iso3, a] = (c_path / a / "updates.h5").exists()
```

```python
get_analysis_status("50038825")
```
