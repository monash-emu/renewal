```python
original_mu = 0.38
sd = 0.1

# original_mu = 0.0339
# sd = 0.0059

# original_mu = 0.0079
# std = 0.0012

adj = 0.4
# 0.3 0.5333333 0.76666667 1.0
```

```python
adj_mu = original_mu * adj
var = sd ** 2
```

```python
def get_beta_params_from_mean_var(mu, var):
    a = mu * (mu * (1.0 - mu) / var - 1.0)
    b = (1.0 - mu) * (mu * (1.0 - mu) / var - 1.0)
    return a, b
```

```python
orig_a, orig_b = get_beta_params_from_mean_var(original_mu, var)
adj_a, adj_b = get_beta_params_from_mean_var(adj_mu, var)
```

```python
x_vals = np.linspace(0.0, 0.5, 500)
orig_vals = beta.pdf(x_vals, orig_a, orig_b)
adj_vals = beta.pdf(x_vals, adj_a, adj_b)
```

```python
pd.DataFrame(
    {
        "orig": orig_vals,
        "adj": adj_vals,
    },
    index=x_vals,
).plot()
```
