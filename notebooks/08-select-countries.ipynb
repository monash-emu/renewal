```python
import pandas as pd
import pycountry
import json

from emu_renewal.inputs import DATA_PATH
```

```python
hosp = pd.read_csv(DATA_PATH / "owid/owid_hosp.csv", index_col="date")
have_occupancy = set(hosp.loc[hosp["indicator"] == "Daily hospital occupancy", "entity"])
have_admits = set(hosp.loc[hosp["indicator"] == "Weekly new hospital admissions", "entity"])
```

```python
eligible_countries = have_occupancy.union(have_admits)
exclude_for_now = ["Australia", "England", "Scotland", "Wales", "Northern Ireland", "Singapore"]
data_issues = ["Iceland", "Liechtenstein", "Bulgaria", "Bolivia", "Cyprus", "Estonia", "Greece", "Croatia", "Hungary", "Ireland", "Japan", "Korea", "Luxembourg", "Latvia", "Malta", "Norway", "Slovenia", "Slovakia", "South Africa"]
all_countries = [c for c in eligible_countries if c not in exclude_for_now + data_issues]
```

```python
initial_countries = {
    "admissions": [pycountry.countries.lookup(c).alpha_2 for c in all_countries if c in have_admits],
    "occupancy": [pycountry.countries.lookup(c).alpha_2 for c in all_countries if c not in have_admits],
}
json.dump(initial_countries, open(DATA_PATH / f"config/countries.json", "w"))
```
