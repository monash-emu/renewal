```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
from matplotlib.pyplot import cm
from os import listdir as ls

from emu_renewal.inputs import OUTPUTS_PATH
from emu_renewal.outputs import get_country_procs
from emu_renewal.plotting import plot_proc_comparison
from emu_renewal.utils import group_countries_by_continent
```

```python
set_matplotlib_formats("svg")
colours = cm.Dark2.colors
```

```python
job_path = OUTPUTS_PATH / "43514097"
countries = ls(job_path)
```

```python
procs = get_country_procs(job_path, countries)
```

```python
continents_map = group_countries_by_continent(countries)
continents_map["EU1"] = continents_map["EU"][:16]
continents_map["EU2"] = continents_map["EU"][16:]
del continents_map["EU"]
```

```python
for continent, countries in continents_map.items():
    plot_proc_comparison(procs, countries, colours, continent, job_path)
```
