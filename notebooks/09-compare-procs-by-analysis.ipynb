```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
from matplotlib.pyplot import cm
from os import listdir as ls
from IPython.display import display, Markdown
import pycountry_convert as pc

from emu_renewal.inputs import OUTPUTS_PATH
from emu_renewal.outputs import get_country_procs
from emu_renewal.plotting import plot_proc_comparison
from emu_renewal.utils import get_countries_by_continent
```

```python
set_matplotlib_formats("svg")
```

```python
job_path = OUTPUTS_PATH / "44255911"
all_countries = ls(job_path)
countries_by_cont = get_countries_by_continent(all_countries)
procs = get_country_procs(job_path, all_countries)
```

```python
for cont in ["EU", "AF", "AS", "NA", "SA", "OC"]:
    cont_name = pc.convert_continent_code_to_continent_name(cont)
    display(Markdown(f"## {cont_name}"))
    n_cols = 3 if cont in ["NA", "SA"] else 4
    if cont == "EU":
        display(plot_proc_comparison(procs, countries_by_cont[cont][:16], cont_name, job_path, n_cols))
        display(plot_proc_comparison(procs, countries_by_cont[cont][16:], cont_name, job_path, n_cols))
    else:
        display(plot_proc_comparison(procs, countries_by_cont[cont], cont_name, job_path, n_cols))
```
