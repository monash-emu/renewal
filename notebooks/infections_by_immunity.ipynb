```python
import copy
from jax import numpy as jnp
import numpy as np
from emu_renewal.renew import get_combs
```

```python
n_strains = 3
strain_map = get_combs(["ba1", "ba2", "ba5"])
actual_inc = jnp.zeros([n_strains, len(strain_map)])
suscept = jnp.zeros(len(strain_map))
```

```python
strain_map
```

```python
for i, dest in enumerate(strain_map):
    dest_cat = strain_map.index(dest)
    for strain in range(n_strains):
        dest[strain] = True
        suscept = suscept.at[dest_cat].set(suscept[dest_cat] + actual_inc[strain, i])
```
