```python
import copy
from jax import numpy as jnp
import numpy as np
from emu_renewal.renew import get_combs
```

```python
n_strains = 3
strain_map = get_combs(["ba1", "ba2", "ba5"])
actual_inc = jnp.zeros([n_strains, len(strain_map)])
suscept = jnp.zeros(len(strain_map))
```

```python
suscept = suscept - actual_inc.sum(axis=0)
for s in range(n_strains):
    for i, imm in enumerate(strain_map):
        dest = copy.copy(imm)
        print(s)
        print(dest)
        dest[s] = True
        print(dest)
        dest_cat = strain_map.index(dest)
        print(dest_cat)
        print("\n")
        suscept = suscept.at[dest_cat].set(suscept[dest_cat] + actual_inc[s, i])
```
