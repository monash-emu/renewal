```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
from os import listdir as ls
from IPython.display import display, Markdown
import pycountry_convert as pc

from emu_renewal.constants import OUTPUTS_PATH
from emu_renewal.plotting import ANALYSIS_NAMES, plot_mob_weights_by_country, get_cont_mobility
from emu_renewal.utils import get_countries_by_continent
```

```python
set_matplotlib_formats("svg")
```

```python
job_path = OUTPUTS_PATH / "44718190"
countries = ls(job_path)
all_countries = ls(job_path)
countries_by_cont = get_countries_by_continent(all_countries)
```

```python
mob_type = "g_mob"
mob_name = ANALYSIS_NAMES["g_mob"]
display(Markdown(f"# {mob_name}"))
for cont in countries_by_cont:
    cont_name = pc.convert_continent_code_to_continent_name(cont)
    display(Markdown(f"## {cont_name}"))
    title = f"Posteriors for {mob_name} domain weight parameters, {cont_name}"
    mob = get_cont_mobility(cont, countries_by_cont, mob_type)
    if cont == "EU":
        eur_countries = list(mob.keys())
        first_countries = {k: mob[k] for k in eur_countries[:16]}
        display(plot_mob_weights_by_country(job_path, mob_type, first_countries, title))
        last_countries = {k: mob[k] for k in eur_countries[16:]}
        display(plot_mob_weights_by_country(job_path, mob_type, last_countries, title))
    else:
        # temp_mob = {k: v for k, v in mob.items() if k != "GNB"}
        display(plot_mob_weights_by_country(job_path, mob_type, mob, title))
```
