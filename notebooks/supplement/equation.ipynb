$$i_{s,t,p}=n_{p}(1-exp(-M_{t}W_{t}\sigma_{s}\sum_{p'}x_{p',s}\sum_{\tau<t}{i_{s,\tau,p'}g_{t-\tau}}))$$
$$M_{t}=
\begin{cases}
1 & \text{if no mobility} \\
(\frac{\sum_{l}w_{l}G_{l,t}}{\sum_{l}w_{l}})^{m} & \text{if Google mobility} \\
F_{t}^{m} & \text{if Facebook mobility} \\
\end{cases}
$$
Where:
- $i_{s,p}$ represents the variant and immunity status-specific incidence
- $p\in \{0, 1\}^{s}$ are the possible combination of past variant exposure statuses
- $n_{p}$ is the size of the population with immunity status $p$
- $W_{t}$ is a sequence of Brownian updates linked with a cosine spline at time $t$
- $\sigma_{s}$ is the relative infectiousness of variant $s$
- $x_{p,s}$ is the relative susceptibility of immunity status $p$ to variant $s$
- $\tau$ and $t$ index time in days
- $g_{t}$ is a discretised gamma distribution
- $G_{l,t}$ is the Google mobility estimate at location $l$ at time $t$
- $F_{t}$ is the Facebook mobility (tiles visited or within tile) at time $t$
- $w_{l}$ is a calibrated location-specific weight parameter
