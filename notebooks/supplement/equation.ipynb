$$i_{v,t,p}=\sum_{p'}n_{t,p'} x_{p',v}(1-exp(-M_{t}e^{W_{t}}\sigma_{v}\sum_{\tau<t}{( s_{v,t} + \sum_{\tilde{p}}i_{v,\tau,\tilde{p}})g_{t-\tau}}))$$
$$n_{t,p}=n_{t-1,p}-\sum_{v}i_{v,t,p}+i_{v',t,p'}$$
$$M_{t}=
\begin{cases}
1 & \text{if no mobility} \\
(\frac{\sum_{l}w_{l}G_{l,t}}{\sum_{l}w_{l}})^{m} & \text{if Google mobility} \\
F_{t}^{m} & \text{if Facebook mobility} \\
\end{cases}
$$
$$
o_{t}=a\sum_{v}\sum_{p}\sum_{t=-50}^{-1}i_{v,t,p}c_{-t}
$$
$$
r_{t}=\sum_{t=-50}^{-1}o_{t}(1-\sum_{\tau}^{-t}d_{-\tau})
$$
Where:
- $i_{v,t,p}$ represents the incidence of variant $v$ at time $t$ in immunity status group $p$
- $p\in \{0, 1\}^{v}$ are the possible combination of past variant exposure statuses
- $s_{v,t}$ is the rate of seeding with variant $v$ at time $t$
- $g_{t}$ is the density of a gamma distribution at time $t$
- $\sigma_{v}$ is the infectiousness of variant $v$ relative to the starting variant
- $W_{t}$ is a Wiener process of updates, linked with a cosine spline at time $t$
- $x_{p,v}$ is the relative susceptibility of persons with immunity status $p$ to infection with variant $v$
- $x_{1,v} = 1$, full susceptibility for those never previously infected
- $x_{p,v} = 0$ for infection with a variant to which the person has previously been infected with or a preceding variant (e.g. infection with Alpha following infection with Delta)
- $x_{p,v}$ for immunity statuses other than the two preceding cases is the complement of a single calibrated cross-immunity parameter with a beta-distributed prior
- $n_{t,p}$ is the size of the population with immunity status $p$ at time $t$
- $G_{l,t}$ is the Google mobility estimate at location $l$ and time $t$
- $F_{t}$ is the Facebook mobility estimate (either tiles visited or single tile) at time $t$
- $w_{l}$ is a calibrated location-specific weight parameter
- $o_{t}$ is any one of the incident indicators calculated from the incidence time series, including deaths, case notifications, hospital admission and ICU admissions
- $c_{t}$ is the distribution of the delay from the onset of the infection epiosde to the onset of the incident indicator considered
- $a$ is the proportion of infection episodes resulting in the indicator considered
- $r_{t}$ is either one of the prevalent indicators calculated from the incidence indicators, including hospital occupancy or ICU occupancy
- $d_{t}$ is the distribution of time from the incident indicator (either hospital or ICU admission) to the end of the prevalent indicator period (either hospital or ICU discharge)
