$$i_{s,t,p}=n_{p}\sum_{p'}x_{p',s}(1-exp(-M_{t}W_{t}\sigma_{s}\sum_{\tau<t}{i_{s,\tau,p'}g_{t-\tau}}))$$
$$M_{t}=
\begin{cases}
1 & \text{if no mobility} \\
(\frac{\sum_{l}w_{l}G_{l,t}}{\sum_{l}w_{l}})^{m} & \text{if Google mobility} \\
F_{t}^{m} & \text{if Facebook mobility} \\
\end{cases}
$$
Where:
- $i_{s,t,p}$ represents the variant-specific, immunity status-specific incidence at time $t$
- $p\in \{0, 1\}^{s}$ are the possible combination of past variant exposure statuses
- $p$ is the immunity status obtained when a person with immunity status $p'$ is infected with variant $s$
- $n_{p}$ is the size of the population with immunity status $p$
- $W_{t}$ is a sequence of Brownian updates linked with a cosine spline at time $t$
- $\sigma_{s}$ is the infectiousness of variant $s$ relative to the starting variant
- $x_{p,s}$ is the relative susceptibility of immunity status $p$ to infection with variant $s$
- $x_{1,s} = 1$, full susceptibility for those never previously infected
- $x_{p,s} = 0$ for infection with a variant to which the person has previously been infected with or a preceding variant (e.g. infection with Alpha following infection with Delta)
- $x_{p,s}$ for immunity statuses other than the two preceding cases is the complement of a single calibrated cross-immunity parameter with a beta-distributed prior
- $g_{t}$ is the density of a gamma distribution at time $t$
- $G_{l,t}$ is the Google mobility estimate at location $l$ and time $t$
- $F_{t}$ is the Facebook mobility estimate (tiles visited or single tile) at time $t$
- $w_{l}$ is a calibrated location-specific weight parameter
