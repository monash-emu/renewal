```python
from os import listdir as ls
from IPython.display import display, Markdown, Latex
import pycountry_convert as pc
from matplotlib_inline.backend_inline import set_matplotlib_formats

from emu_renewal.constants import OUTPUTS_PATH, CONT_CMAP
from emu_renewal.plotting import compare_proc_pol
from emu_renewal.utils import get_countries_by_continent, get_job_commits_df, split_list_into_segments
from emu_renewal.selection import find_pol_countries

set_matplotlib_formats("svg")
```

```python
job_path = OUTPUTS_PATH / "47213985"
all_countries = ls(job_path)
pol_countries = find_pol_countries(all_countries)
countries_by_cont = get_countries_by_continent(pol_countries)
```

```python
display(Markdown("# Policy profile comparisons \n\n"))
display(Latex(r"\newpage"))
for cont in CONT_CMAP:
    countries = countries_by_cont[cont]
    display(Markdown(f"### {pc.convert_continent_code_to_continent_name(cont)}"))
    country_groups = split_list_into_segments(countries, 9)
    for group in country_groups:
        display(compare_proc_pol(job_path, group, 3))
```
