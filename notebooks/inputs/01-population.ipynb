The following code was used to process the data downloaded from: https://population.un.org/wpp/assets/Excel%20Files/1_Indicator%20(Standard)/EXCEL_FILES/2_Population/WPP2024_POP_F02_1_POPULATION_5-YEAR_AGE_GROUPS_BOTH_SEXES.xlsx
on 18th March 2025.
This created the file with the UN DESA population data (called by get_undesa_national_pop).

```python
import pandas as pd
from emu_renewal.inputs import DATA_PATH
```

```python
pop_year = 2020
filename = "population/WPP2024_POP_F02_1_POPULATION_5-YEAR_AGE_GROUPS_BOTH_SEXES.xlsx"
all_data = pd.read_excel(DATA_PATH / filename, skiprows=16, index_col=0)
all_data["population"] = all_data.iloc[:, -21:].sum(axis=1)
pop_data = all_data[["ISO3 Alpha-code", "Year", "population"]].dropna()
year_pop_data = pop_data[pop_data["Year"] == float(pop_year)]
final_pops = year_pop_data["population"]
final_pops.index = year_pop_data["ISO3 Alpha-code"]
final_pops.to_csv(DATA_PATH / f"population/undesa_pops_{pop_year}.csv")
```
