```python
import json

from emu_renewal.inputs import DATA_PATH, process_raw_google_mobility
```

```python
countries = json.load(open(DATA_PATH / "config/included.json", "r"))
for c in countries:
    try:
        data = process_raw_google_mobility(c)
        data.to_csv(DATA_PATH / f"mobility/{c}_gmob_data.csv")
        print(f"extracted g_mob data for {c}")
    except:
        f"no g_mob data for {c}"
```
