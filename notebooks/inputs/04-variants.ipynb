The following code was run on 11 December 2024 to generate the stored Nextclade genomics data CSV file.

```python
import pandas as pd
from os import listdir as ls

from emu_renewal.inputs import DATA_PATH
```

```python
var_names = [v.split(".json")[0] for v in ls(DATA_PATH / "nextclade") if v.startswith("2")]
for var in var_names:
    path = DATA_PATH / f"nextclade/{var}.json"
    if path.exists():
        print(f"Already extracted {var}")
    else:
        print(f"Extracting {var}")
        url = f"https://github.com/hodcroftlab/covariants/raw/refs/heads/master/cluster_tables/{var}_data.json"
        json_data = pd.read_json(url)
        json_data.to_json(DATA_PATH / f"nextclade/{var}.json")
```
