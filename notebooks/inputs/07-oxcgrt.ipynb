```python
import pandas as pd

dtypes = {
    "V2B_Vaccine age eligibility/availability age floor (general population summary)": str,
    "V2C_Vaccine age eligibility/availability age floor (at risk summary)": str,
    "V4_Notes": str,
    "M1_Notes": str,
}
sheets = []
for year in range(2020, 2023):
    url = f"https://github.com/OxCGRT/covid-policy-dataset/raw/refs/heads/main/data/OxCGRT_fullwithnotes_national_{year}_v1.csv"
    sheets.append(pd.read_csv(url, dtype=dtypes))
```

```python
data = pd.concat(sheets)
data.index = pd.to_datetime(data["Date"], format="%Y%m%d")
```

```python
iso3 = "ABW"
abw_data = data[data["CountryCode"] == iso3]
```

```python
abw_data["StringencyIndex_SimpleAverage"].plot()
```
