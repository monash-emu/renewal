The following code was used to process the data downloaded from: the URL listed below on 14th September 2025.

```python
import pandas as pd

from emu_renewal.constants import OXCGRT_DTYPES, OXCGRT_IND_MAX
from emu_renewal.inputs import DATA_PATH, get_oxcgrt, get_oxcgrt_country_indicators, get_rel_oxcgrt_cols, scale_oxcgrt_pols
```

```python
sheets = []
for year in range(2020, 2023):
    url = f"https://github.com/OxCGRT/covid-policy-dataset/raw/refs/heads/main/data/OxCGRT_fullwithnotes_national_{year}_v1.csv"
    sheets.append(pd.read_csv(url, dtype=OXCGRT_DTYPES))
data = pd.concat(sheets)
data.to_csv(DATA_PATH / f"restrictions/oxcgrt.csv")
```

```python
stringency_cols = [f"C{i}" for i in range(1, 9)] + ["H1"]
mob_pol_cols = [f"C{i}" for i in range(1, 8)] + ["H1"]
```

```python
pol = get_oxcgrt_country_indicators("GBR")
filt_pol = pol[get_rel_oxcgrt_cols("M", pol)]
scaled_pol = scale_oxcgrt_pols(filt_pol)
scaled_pol[mob_pol_cols].mean(axis=1).plot()
```
