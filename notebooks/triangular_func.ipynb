```python
import numpy as np
import pandas as pd
```

```python
start = 10.0
end = 20.0
peak = 30.0
```

```python
def triangular_func(val):
    half_width = (end - start) / 2.0
    dist_from_midpoint = abs(val - (start + end) / 2.0)
    vals = 1.0 - dist_from_midpoint / half_width
    vals[vals < 0.0] = 0.0
    return vals * peak
```

```python
x_vals = np.linspace(0.0, 50.0, 1000)
result = triangular_func(x_vals)
```

```python
pd.Series(result, index=x_vals).plot()
```
