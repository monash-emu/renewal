```python
from datetime import datetime
import numpy as np
from numpy.random import normal
import pandas as pd
import plotly.express as px

from emu_renewal.renew import SimpleModel
from emu_renewal.mobility import NoMobilityProvider

pd.options.plotting.backend = "plotly"
```

```python
pop = 1e6
run_start = datetime(2020, 1, 1)
end_time = datetime(2020, 12, 31)
mob_provider = NoMobilityProvider()
model = SimpleModel(
    pop,
    run_start,
    end_time,
    mob_provider,
)
```

```python
out = model.renewal_func(
    1.1,
    normal(scale=0.05, size=len(model.x_proc_data.points)),
    3.0,
    1.0,
    0.5,
    0.01,
    5.0,
    1.0,
    20.0,
    2.0,
    -24.0,
)
pd.Series(out["inc"]).plot()
```

```python
px.scatter(x=model.model_times[::7], y=out["inc"][::7])
```
