```python
import matplotlib.pyplot as plt
import geopandas as gpd
import json
from emu_renewal.inputs import DATA_PATH
```

```python
included = json.load(open(DATA_PATH / "config/included.json", "r"))
excluded = json.load(open(DATA_PATH / "config/excluded.json", "r"))
```

```python
world = gpd.read_file(DATA_PATH / "mapping/ne_10m_admin_0_countries.shp")
world["included"] = world["ADM0_ISO"].isin(included)
world["excluded"] = world["ADM0_ISO"].isin(excluded)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, color="black", linewidth=0.2)
world[world["included"]].plot(ax=ax, color="darkblue")
world[world["excluded"]].plot(ax=ax, color="red")
```
