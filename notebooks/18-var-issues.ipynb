```python
import pycountry
from emu_renewal.inputs import get_alpha_target, get_country_vacc_data, get_country_vars, get_delta_target, get_var_target, extract_specific_var, get_continent_data, get_continent_vars
from emu_renewal.run import find_run_end_time
import pycountry
import pycountry_convert as pc

country = pycountry.countries.lookup("LBN")
iso3 = country.alpha_3
country
```

```python
delta_data = get_delta_target(var_data, iso3, continent, end_time)
```

```python
if continent not in ["OC", "AF"]:
    alpha_prop = get_alpha_target(var_data, iso3, continent, end_time, delta_data)
    alpha_prop.plot()
```
