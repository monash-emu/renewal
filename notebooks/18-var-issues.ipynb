```python
import pycountry
from emu_renewal.inputs import get_alpha_target, get_country_vacc_data, get_country_vars, get_delta_target
from emu_renewal.run import find_run_end_time
import pycountry_convert as pc

country = pycountry.countries.lookup("CIV")
iso3 = country.alpha_3
country
```

```python
vacc_data = get_country_vacc_data(iso3)
var_data = get_country_vars(iso3)
iso2 = pycountry.countries.lookup(iso3).alpha_2
continent = pc.country_alpha2_to_continent_code(iso2)
end_time = find_run_end_time(vacc_data, 0.05, iso3)
delta_data = get_delta_target(var_data, continent, end_time)
alpha_prop = get_alpha_target(var_data, continent, end_time, delta_data)
```
