```python
import pandas as pd
import pickle
import arviz as az
from emu_renewal.outputs import get_output_dir, load_targets
from emu_renewal.plotting import plot_spaghetti_calib_comparison, plot_post_prior_comparison
```

```python
country = "France"
analysis = "no_mob"
time = "16122024_1942"
```

```python
spaghetti = pd.read_hdf(get_output_dir(country, analysis, time) / "spaghetti.h5")
targets = load_targets(country, analysis, time)
idata = az.from_netcdf(get_output_dir(country, analysis, time) / "idata.nc")
```

```python
plot_spaghetti_calib_comparison(spaghetti, targets, list(targets.keys()) + ["process"])
```

```python
priors = pickle.load(open(get_output_dir(country, analysis, time) / "priors.pkl", 'rb'))
```

```python
plot_post_prior_comparison(idata, ["cdr", "stay_mean", "stay_sd", "ifr", "har"], priors);
```
