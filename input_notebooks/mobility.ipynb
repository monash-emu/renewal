The following code was run on 14/1/25 to generate the  Google Community Mobility data CSV files.
The original CSV files from Google were obtained from https://www.gstatic.com/covid19/mobility/Region_Mobility_Report_CSVs.zip
on 11/12/24. These were saved to a folder ("mobility_raw") which has not been pushed to this repo because of the size of the files.

```python
import pandas as pd
import pycountry
from emu_renewal.inputs import DATA_PATH, get_country_mobility
```

```python
countries = ["France", "Spain", "Italy"]
for c in countries:
    data = get_country_mobility(c)
    iso2 = pycountry.countries.get(name=c).alpha_2
    data.to_csv(DATA_PATH / f"mobility/{iso2}_mob_data.csv")
```
