The following code was run on 19/11/24 to generate the Australian Google Community Mobility data CSV file.
The original CSV files from Google were obtained from https://www.gstatic.com/covid19/mobility/Region_Mobility_Report_CSVs.zip
on 19/11/24.

```python
from pathlib import Path
import pandas as pd

PROJECT_PATH = Path.cwd().resolve()
DATA_PATH = PROJECT_PATH.parent / "data/covid_aus"
```

```python
mob_data_2020 = pd.read_csv(DATA_PATH / "2020_AU_Region_Mobility_Report.csv", index_col="date")
mob_data_2021 = pd.read_csv(DATA_PATH / "2021_AU_Region_Mobility_Report.csv", index_col="date")
mob_data_2022 = pd.read_csv(DATA_PATH / "2022_AU_Region_Mobility_Report.csv", index_col="date")
```

```python
mob_data = pd.concat([mob_data_2020, mob_data_2021, mob_data_2022])
```

```python
mob_data.index = pd.to_datetime(mob_data.index)
```

```python
aust_data = mob_data.loc[pd.isna(mob_data["sub_region_1"])]
aust_data = aust_data[[c for c in aust_data.columns if "change_from_baseline" in c]]
aust_data = aust_data.rename(lambda x: x.replace("_percent_change_from_baseline", ""), axis=1)
aust_data = 1.0 + aust_data / 100.0
```

```python
aust_data.to_csv(DATA_PATH / "aust_mobility.csv")
```
