The following code was run on 11/12/24 to generate the  Google Community Mobility data CSV files.
The original CSV files from Google were obtained from https://www.gstatic.com/covid19/mobility/Region_Mobility_Report_CSVs.zip
on 11/12/24.

```python
from pathlib import Path
import pandas as pd
from emu_renewal.inputs import PROJECT_PATH, DATA_PATH
```

```python
DATA_PATH = DATA_PATH / "covid_aus"
```

```python
def get_country_mobility(country):
    mob_data_2020 = pd.read_csv(DATA_PATH / f"2020_{country}_Region_Mobility_Report.csv", index_col="date")
    mob_data_2021 = pd.read_csv(DATA_PATH / f"2021_{country}_Region_Mobility_Report.csv", index_col="date")
    mob_data_2022 = pd.read_csv(DATA_PATH / f"2022_{country}_Region_Mobility_Report.csv", index_col="date")
    data = pd.concat([mob_data_2020, mob_data_2021, mob_data_2022])
    data.index = pd.to_datetime(data.index)
    national_data = data.loc[pd.isna(data["sub_region_1"])]
    mob_columns = [c for c in data.columns if "change_from_baseline" in c]
    national_data = national_data[mob_columns]
    national_data = national_data.rename(lambda c: c.replace("_percent_change_from_baseline", ""), axis=1)
    national_data = 1.0 + national_data / 100.0
    return national_data
```

```python
get_country_mobility("AU")
```
