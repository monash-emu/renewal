The following code was run on 11/12/24 to generate the Nextclade genomics data CSV file.

```python
import pandas as pd
from emu_renewal.inputs import DATA_PATH, VAR_MAP
```

```python
for var in VAR_MAP.values():
    url = f"https://github.com/hodcroftlab/covariants/raw/refs/heads/master/cluster_tables/{var}_data.json"
    json_data = pd.read_json(url)
    json_data.to_json(DATA_PATH / "nextclade" / f"{var}.json")
```
