The following code was run on 11/12/24 to generate the Nextclade genomics data CSV file.

```python
import pandas as pd
from datetime import datetime
import re

from emu_renewal.inputs import DATA_PATH, VAR_NAMES
```

```python
for var in VAR_NAMES:
    path = DATA_PATH / f"nextclade/{var}.json"
    if path.exists():
        print(f"Already extracted {var}")
    else:
        print(f"Extracting {var}")
        url = f"https://github.com/hodcroftlab/covariants/raw/refs/heads/master/cluster_tables/{var}_data.json"
        json_data = pd.read_json(url)
        json_data.to_json(DATA_PATH / f"nextclade/{var}.json")
```

```python
pd.options.plotting.backend = "plotly"
```

```python
country_all_var_data = pd.DataFrame()
country = "France"
for var in VAR_NAMES:
    var_data = pd.read_json(DATA_PATH / f"nextclade/{var}.json")
    if country in var_data:
        country_var_data = var_data[country]
        country_all_var_data[var] = pd.Series(country_var_data["cluster_sequences"], index=pd.to_datetime(country_var_data["week"]))
```

```python
country_all_var_data.plot.area()
```

```python
relevant_vars = []
for var in country_all_var_data:
    if country_all_var_data.loc[country_all_var_data.index < datetime(2021, 2, 1), var].sum() > 100:
        relevant_vars.append(var)
```

```python
country_all_var_data[relevant_vars].plot.area()
```
