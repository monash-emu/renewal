```python
import pandas as pd
from emu_renewal.inputs import DATA_PATH
```

```python
filename = "population/WPP2024_POP_F02_1_POPULATION_5-YEAR_AGE_GROUPS_BOTH_SEXES.xlsx"
all_data = pd.read_excel(DATA_PATH / filename, skiprows=16, index_col=0)
all_data["population"] = all_data.iloc[:, -21:].sum(axis=1)
pop_data = all_data[["ISO3 Alpha-code", "Year", "population"]].dropna()
total_pops = pop_data["population"]
total_pops.index = pd.MultiIndex.from_arrays([pop_data["ISO3 Alpha-code"], pop_data["Year"].astype(int)])
total_pops *= 1e3
```

```python
pop_year = 2020

year_pops = total_pops[total_pops.index.isin([pop_year],level=1)].droplevel(1)
year_pops.to_csv(f"undesa_pops_{pop_year}.csv")
```
