Standard workflow for syncing is to run the code below and then `aws s3 sync .\sync\ s3://autumn-data/projects/renewal/outputs/`

```python
import shutil
from emu_renewal.inputs import BASE_PATH, ANALYSIS_TYPES
from emu_renewal.outputs import get_latest_analyses

for country in ["France", "Italy", "Portugal", "Spain", "Sweden"]:
    country_analysis_ids = get_latest_analyses(BASE_PATH / "outputs" / country, ANALYSIS_TYPES)
    for analysis, time in country_analysis_ids.items():
        analysis_subpath = f"{country}/{analysis}/{time}"
        
        source = BASE_PATH / "outputs" / analysis_subpath
        dest = BASE_PATH / "sync" / analysis_subpath
        if not dest.exists():
            shutil.copytree(source, dest)
```
